```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering


patient1 = [76,126,38.0]
patient2 =[74,120,38.0]
patient3 = [72,118,37.5]
patient4 = [78,136,37.0]


patients = [patient1, patient2, patient3, patient4]
data = np.array(patients)
print(patients)

distances = np.linalg.norm(data[:, np.newaxis] - data, axis=-1)
```

```python
farthest_patient_idx = np.unravel_index(np.argmax(distances), distances.shape)
farthest_patient = farthest_patient_idx[0]

print("The farthest patient index:", farthest_patient)
```

```python
np.fill_diagonal(distances, np.inf)

patient1_index, patient2_index = np.unravel_index(np.argmin(distances), distances.shape)
min_distance = distances[patient1_index, patient2_index]


print(f"The two nearest patients are Patient {patient1_index + 1} and Patient {patient2_index + 1}")
```

```python
new_dummy_patient = np.array([75, 130, 37.8])

new_patient_distances = np.linalg.norm(data - new_dummy_patient, axis=1)

closest_patient_index = np.argmin(new_patient_distances)

print(f"The closest patient to the new dummy patient is Patient {closest_patient_index + 1}")
```
